```python

query = """
WITH RankedPokemon AS (
  SELECT
    Type_1,
    Name,
    Total,
    RANK() OVER (PARTITION BY Type_1 ORDER BY Total DESC) as Rank
  FROM
    PokemonStats
)
SELECT
  Type_1,
  Name,
  Total
FROM
  RankedPokemon
WHERE
  Rank <= 3;
"""

# Execute the query and load into a DataFrame
results_df = pd.read_sql_query(query, engine)

# Display the results
results_df
```
